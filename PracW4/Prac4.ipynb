```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Define the function
def f(x):
    return x**3 + 1

# Generate the x values
x = np.linspace(-1, 1, 100)

# Generate the y values by applying the function to the x values
y = f(x)

# Plot the function
plt.plot(x, y)

# Add labels and a title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function plot')

# Show the plot
plt.show()
```

```python
# Generate 30 random sample training set
x_train = np.random.uniform(-1, 1, 30)
noise = np.random.normal(0,0.5,30)
y_test = f(x_train)
y_train = f(x_train) + noise
x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
y_train.shape
```

```python
# Plot the training set
plt.scatter(x_train, y_train)

# Add labels and a title
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.title('training plot')

# Show the plot
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred = lin_reg.predict(x_train)
y_pred.shape
```

```python
def plot_linear(x_train, y, y_pred):
    # Plot the function
    plt.plot(x_train, y_pred)
    plt.scatter(x_train, y, color='r', label='Data Points')

    # Add labels and a title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function plot')

    # Show the plot
    plt.show()
    
plot_linear(x_train, y_train, y_pred)
```

```python
sum_of_square_error = np.sum((y_test - y_pred) ** 2)
sum_of_square_error
```

```python
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X2 = poly_features.fit_transform(x.reshape(-1,1))
poly_reg_model = LinearRegression()
poly_reg_model.fit(X2, y)
y2_pred = poly_reg_model.predict(X2)
y2_pred.shape
plt.plot(x, y, color='r')
```

```python
def plot_poly(x, y, x_ploy, y_pred):
    # Plot the function
    plt.plot(x_ploy, y_pred, color='r')
#     plt.scatter(x, y, color='r', label='Data Points')

    # Add labels and a title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function plot')

    # Show the plot
    plt.show()
    
plot_poly(x_train, y_train, X2, y_test)
```
